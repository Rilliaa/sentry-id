```python
# =========================================
# 01_eda.ipynb - EDA untuk Sentry-ID Level 1
# =========================================

# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Path dataset
import pandas as pd
data_path = "/content/drive/My Drive/Colab Notebooks/Sentry-ID/Dataset/dataset.csv"

# Load dataset
df = pd.read_csv(data_path)

# Cek 5 baris awal
print("Preview dataset:")
display(df.head())

# Info dasar
print("\nInfo dataset:")
print(df.info())

print("\nJumlah data per kelas:")
print(df['label'].value_counts())

print("\nProporsi data:")
print(df['label'].value_counts(normalize=True))

# =========================================
# Statistik teks
# =========================================
df['char_len'] = df['comment'].astype(str).apply(len)
df['token_len'] = df['comment'].astype(str).apply(lambda x: len(x.split()))

print("\nStatistik panjang teks per kelas:")
display(df.groupby('label')[['char_len', 'token_len']].describe())

# =========================================
# Visualisasi distribusi
# =========================================
import matplotlib.pyplot as plt
import seaborn as sns

# Distribusi kelas
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='label', order=df['label'].value_counts().index, palette="viridis")
plt.title("Distribusi Kelas")
plt.show()

# Distribusi panjang teks
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='token_len', hue='label', multiple='stack', bins=30)
plt.title("Distribusi Panjang Komentar (token)")
plt.xlabel("Jumlah Token")
plt.show()

# =========================================
# Word Frequency / TF-IDF sederhana
# =========================================
from sklearn.feature_extraction.text import CountVectorizer

def get_top_n_words(corpus, n=15):
    vec = CountVectorizer(stop_words=None).fit(corpus)
    bag = vec.transform(corpus)
    sum_words = bag.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    return sorted(words_freq, key=lambda x: x[1], reverse=True)[:n]

for label in df['label'].unique():
    subset = df[df['label'] == label]
    top_words = get_top_n_words(subset['comment'].astype(str), 15)
    print(f"\nTop kata untuk kelas {label}:")
    for word, freq in top_words:
        print(f"{word}: {freq}")

# =========================================
# Wordcloud per kelas (opsional)
# =========================================
from wordcloud import WordCloud

for label in df['label'].unique():
    subset = df[df['label'] == label]
    text = " ".join(subset['comment'].astype(str))
    wc = WordCloud(width=800, height=400, background_color="white").generate(text)
    plt.figure(figsize=(10,5))
    plt.imshow(wc, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"Wordcloud untuk kelas {label}")
    plt.show()
```
